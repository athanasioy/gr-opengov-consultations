```python
from textUtils.textDecorators import TextDecoratorChainFactory

decoratorChain = TextDecoratorChainFactory.createDecoratorChain(applyLower=False)

with Session(engine) as sess:
    stmt = select(Article)
    row_count_stmt = select(func.count(Article.id))
    row_count = sess.execute(row_count_stmt).scalar()
    results = sess.execute(stmt).scalars()
    print(row_count)
    for idx,articleObj in enumerate(results):
        # articleObj.text = decoratorChain.execute(articleObj.text)
        articleObj.title = decoratorChain.execute(articleObj.title)
        

        if idx %100 == 0:
            print(f"Progress: {idx/row_count:.2%}")

    sess.commit()
```
